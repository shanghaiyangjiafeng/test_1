<h2 style='background: black; border:1; color: white'><center>2.2 模型计算</center></h2>

```python
#LMDI法 
def LMDI(data,tt,t0):
    def Delta_XX(yt,y0,xt,x0):
        def L(yt,y0):
            if yt == y0:
                return 0
            else:
                return (yt-y0)/(np.log(yt) - np.log(y0))
        return L(yt,y0)*np.log(xt/x0)
    Delta_y = data[tt,0]-data[t0,0]
    Delta_x = []
    for i in range(1,np.shape(data)[1]):
        Delta_x.append( Delta_XX(data[tt,0],data[t0,0],data[tt,i], data[t0,i]) )
    return Delta_y,Delta_x

#分解
def dt(b,d):
    for i in range(0,np.shape(b)[0]-1):
        c_0=LMDI(b,i+1,i)
        c_1=np.array(c_0)
        c_2=np.array(c_1[1])
        d[i]=c_2[0]
    return d

a_1=pd.read_excel(r'3_1_1.xlsx')
a_2=pd.read_excel(r'3_1_2.xlsx')
a_3=pd.read_excel(r'3_1_3.xlsx')
a_4=pd.read_excel(r'3_1_4.xlsx')
a_5=pd.read_excel(r'3_1_5.xlsx')
a_6=pd.read_excel(r'3_1_6.xlsx')
a_7=pd.read_excel(r'3_1_7.xlsx')

a=a_1.iloc[:,1:3]
b=a.values
d=[0 for i in range(np.shape(b)[0]-1)]
b_1=dt(b,d)

a=a_2.iloc[:,1:5]
b=a.values
d=[0 for i in range(np.shape(b)[0]-1)]
b_2=dt(b,d)

a=a_3.iloc[:,1:4]
b=a.values
d=[0 for i in range(np.shape(b)[0]-1)]
b_3=dt(b,d)

a=a_4.iloc[:,1:4]
b=a.values
d=[0 for i in range(np.shape(b)[0]-1)]
b_4=dt(b,d)

a=a_5.iloc[:,1:4]
b=a.values
d=[0 for i in range(np.shape(b)[0]-1)]
b_5=dt(b,d)

a=a_6.iloc[:,1:3]
b=a.values
d=[0 for i in range(np.shape(b)[0]-1)]
b_6=dt(b,d)

a=a_7.iloc[:,1:4]
b=a.values
d=[0 for i in range(np.shape(b)[0]-1)]
b_7=dt(b,d)
```
